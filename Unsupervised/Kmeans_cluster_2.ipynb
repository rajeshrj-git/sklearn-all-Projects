# Customer Segmentation using KMeans Alg

```python
from sklearn.cluster import KMeans
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
X = df.iloc[:,[3,4]].values
X
```

# KMeans 

```python
k = 3

kmeans = KMeans(n_clusters=k, random_state=0)
Y = kmeans.fit_predict(X)

centroids = kmeans.cluster_centers_

for i in range(k):
    plt.scatter(X[Y == i, 0], X[Y == i, 1], label=f'Cluster {i+1}')

plt.scatter(centroids[:, 0], centroids[:, 1], s=100, c='black', label='Centroids', marker='X')

plt.title('K-means Clustering')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.show()
print(centroids)
Y
```

# Tuning Methods Rugularizations

```python
from sklearn.cluster import MiniBatchKMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score



scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


inertia = []
K = range(1, 11)
for k in K:
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=0)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

print(f"\n Elbow  Values : {inertia}")

plt.figure(figsize=(8, 4))
plt.plot(K, inertia, 'bx-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method  optimal k')
plt.show()


optimal_k = 3

kmeans = KMeans(n_clusters=optimal_k, init='k-means++', random_state=0)
Y = kmeans.fit_predict(X)
centroids = kmeans.cluster_centers_

silhouette_avg = silhouette_score(X_scaled, Y)
print(f' \n Silhouette Score: {silhouette_avg}')

# mbk = MiniBatchKMeans(n_clusters=optimal_k, init='k-means++', random_state=0, batch_size=100)
# Y_mbk = mbk.fit_predict(X_scaled)

pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X_scaled)

kmeans_pca = KMeans(n_clusters=optimal_k, init='k-means++', random_state=0)
Y_pca = kmeans_pca.fit_predict(X_reduced)
centroids_pca = kmeans_pca.cluster_centers_

plt.figure(figsize=(8, 4))
for i in range(optimal_k):
    plt.scatter(X[Y == i, 0], X[Y == i, 1], label=f'Cluster {i+1}')
plt.scatter(centroids[:, 0], centroids[:, 1], s=100, c='black', label='Centroids', marker='X')
plt.title('K-means Clustering')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.show()

plt.figure(figsize=(8, 4))
for i in range(optimal_k):
    plt.scatter(X_reduced[Y_pca == i, 0], X_reduced[Y_pca == i, 1], label=f'Cluster {i+1}')
plt.scatter(centroids_pca[:, 0], centroids_pca[:, 1], s=100, c='black', label='Centroids', marker='X')
plt.title('K-means Clustering with PCA')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.show()

Y
```
